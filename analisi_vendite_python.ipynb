```python
df_orders = pd.read_csv("C:\\Users\\jouin\\Downloads\\archive\\orders (1).csv")
df_orders.head()
```

```python
df_products = pd.read_csv("C:\\Users\\jouin\\Downloads\\archive\\products (1).csv")
df_products.head()
```

```python
df_merged = df_orders.merge(df_products, on="product_id")
df_merged.groupby("product_name")["quantity"].sum().sort_values(ascending=False)
```

```python
df_merged = df_orders.merge(df_products, on="product_id")
df_merged["fatturato"] = df_merged["quantity"] * df_merged["price"]
df_merged.groupby("category")["fatturato"].sum().sort_values(ascending=False)
```
